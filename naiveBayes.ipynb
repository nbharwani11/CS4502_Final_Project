```python
times = []
dates = []
cats = []
neighborhoods = []
hours = []
iscrime = []
for col, row in df.iterrows():
    temp_t = (row["FIRST_OCCURRENCE_DATE"])[-5:]
    temp_d = (row["FIRST_OCCURRENCE_DATE"])[:-5]
    temp_c = (row["OFFENSE_CATEGORY_ID"])
    temp_n = (row["NEIGHBORHOOD_ID"])
    temp_i = (row["IS_CRIME"])
    if temp_t[0] == " ":
        temp_t = "0" + temp_t[1:]
    temp_h = temp_t[:2] 
    times.append(temp_t[:2] + temp_t[-2:])
    dates.append(temp_d)
    cats.append(temp_c)
    neighborhoods.append(temp_n)
    hours.append(temp_h)
    iscrime.append(temp_i)
print(len(times))
print(len(dates))
print(len(cats))
print(len(neighborhoods))
print(len(hours))
print(len(iscrime))
```

```python
from sklearn import preprocessing 
labeler = preprocessing.LabelEncoder()

ret_c = labeler.fit_transform(cats)
ret_n = labeler.fit_transform(neighborhoods)
ret_h = labeler.fit_transform(hours)
print(len(ret_c))
print(len(ret_n))
print(len(ret_h))
```
